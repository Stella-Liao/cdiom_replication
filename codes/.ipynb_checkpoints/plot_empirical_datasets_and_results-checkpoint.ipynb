```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
grouped_o = od_df.groupby('origin_geocode')['flow'].sum().reset_index()
grouped_o.columns = ['geocode', 'outflow']

grouped_d = od_df.groupby('destination_geocode')['flow'].sum().reset_index()
grouped_d.columns = ['geocode', 'inflow']

cnty_gdf2 = cnty_gdf.copy()
cnty_gdf2 = cnty_gdf2.merge(grouped_o, on='geocode', how='left')   
cnty_gdf2 = cnty_gdf2.merge(grouped_d, on='geocode', how='left')
# cnty_gdf2['pop'] = cnty_gdf2['pop']/1000000
```

```python
cnty_gdf2
```

```python
cnty_gdf2 = cnty_gdf2.to_crs("EPSG:4326")
cnty_gdf2
```

```python
cnty_gdf2['pop'] = np.log(cnty_gdf2['pop'] )
cnty_gdf2['unemp'] = np.log(cnty_gdf2['unemp'] )
cnty_gdf2['income'] = np.log(cnty_gdf2['income'] )
cnty_gdf2['temp'] = np.log(cnty_gdf2['temp'] )
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 12))
axes = axes.flatten()

ax = axes[0]
divider = make_axes_locatable(ax)
cax = divider.append_axes("bottom", size="5%", pad=0.5)

# Plotting the data with legend options specified directly
plot = cnty_gdf2.plot(column='pop', ax=ax,
                cmap='Greys',edgecolor='grey', lw=0.3,
                legend=True, cax = cax)

cbar = fig.colorbar(plot.collections[0], cax=cax, orientation='horizontal')
cbar.set_label('Ln(Population) in 2018', fontsize=20) 
cbar.ax.tick_params(labelsize=16)
ax.tick_params(axis='both', which='major', labelsize=18)

ax = axes[1]
divider = make_axes_locatable(ax)
cax = divider.append_axes("bottom", size="5%", pad=0.5)

# Plotting the data with legend options specified directly
plot = cnty_gdf2.plot(column='unemp', ax=ax,
                cmap='Greys',edgecolor='grey', lw=0.3,
                legend=True, cax = cax)

cbar = fig.colorbar(plot.collections[0], cax=cax, orientation='horizontal')
cbar.set_label('Ln(Unemployment rates) in 2018 (%)', fontsize=20) 
cbar.ax.tick_params(labelsize=16)
ax.tick_params(axis='both', which='major', labelsize=18)

ax = axes[2]
divider = make_axes_locatable(ax)
cax = divider.append_axes("bottom", size="5%", pad=0.5)

# Plotting the data with legend options specified directly
plot = cnty_gdf2.plot(column='income', ax=ax,
                cmap='Greys',edgecolor='grey', lw=0.3,
                legend=True, cax = cax)

cbar = fig.colorbar(plot.collections[0], cax=cax, orientation='horizontal')
# cbar.set_label('Ln(Median household income) in 2018 \n (in 2018 inflation-adjusted dollars)', fontsize=20) 
cbar.set_label('Ln(Median household income) in 2018', fontsize=20) 
cbar.ax.tick_params(labelsize=16)
ax.tick_params(axis='both', which='major', labelsize=18)

ax = axes[3]
divider = make_axes_locatable(ax)
cax = divider.append_axes("bottom", size="5%", pad=0.5)

# Plotting the data with legend options specified directly
plot = cnty_gdf2.plot(column='temp', ax=ax,
                cmap='Greys',edgecolor='grey', lw=0.3,
                legend=True, cax = cax)

cbar = fig.colorbar(plot.collections[0], cax=cax, orientation='horizontal')
cbar.set_label('Ln(Annual mean temperature) in 2018 (Â°F)', fontsize=20) 
cbar.ax.tick_params(labelsize=16)
ax.tick_params(axis='both', which='major', labelsize=18)

plt.tight_layout()
plt.savefig('pics/ln_alls.png', dpi=300, bbox_inches='tight')
plt.show() 
```

# Moran's I

```python
# Moran's I & Moran's I Plot
from esda import Moran, Moran_Local 
from libpysal.weights import DistanceBand
from scipy.spatial.distance import pdist, squareform
import numpy as np
```

```python
cnty_gdf3 = cnty_gdf.copy()
cnty_gdf3['centroid'] = cnty_gdf3.geometry.centroid
cnty_gdf3 = cnty_gdf3.set_geometry(cnty_gdf3.geometry.centroid)

centroid_coords = np.array(list(cnty_gdf3['centroid'].apply(lambda x: (x.x, x.y))))
pairwise_distances = pdist(centroid_coords)  # Computes distances in condensed form
average_distance = pairwise_distances.mean()

thresh = average_distance
print(f"Average county size: {average_county_size:.2f} meters")
```

```python
cnty_gdf3['pop'] = np.log(cnty_gdf3['pop'] )
cnty_gdf3['unemp'] = np.log(cnty_gdf3['unemp'] )
cnty_gdf3['income'] = np.log(cnty_gdf3['income'] )
cnty_gdf3['temp'] = np.log(cnty_gdf3['temp'] )
```

```python
w_distance = DistanceBand.from_dataframe(cnty_gdf3, threshold=thresh)
w_distance.transform = 'r'
moran_i_pop = Moran(cnty_gdf3['pop'], w_distance)
moran_i_pop.I, moran_i_pop.z_sim, moran_i_pop.p_sim
```

```python
w_distance = DistanceBand.from_dataframe(cnty_gdf3, threshold=thresh)
w_distance.transform = 'r'
cnty_gdf3['unemp'] = pd.to_numeric(cnty_gdf3['unemp'])
moran_i_unemp = Moran(cnty_gdf3['unemp'], w_distance)
moran_i_unemp.I, moran_i_unemp.z_sim, moran_i_unemp.p_sim
```

```python
w_distance = DistanceBand.from_dataframe(cnty_gdf3, threshold=thresh)
w_distance.transform = 'r'
moran_i_inc = Moran(cnty_gdf3['income'], w_distance)
moran_i_inc.I, moran_i_inc.z_sim, moran_i_inc.p_sim
```

```python
w_distance = DistanceBand.from_dataframe(cnty_gdf3, threshold=thresh)
w_distance.transform = 'r'
moran_i_temp = Moran(cnty_gdf3['temp'], w_distance)
moran_i_temp.I, moran_i_temp.z_sim, moran_i_temp.p_sim
```

# Spatial distribution of smoothed population

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
from terms import SATerm
```

```python
cnty_gdf = gpd.read_file('data/cnty_2019_18.geojson')
cnty_gdf['geocode'] = cnty_gdf['geocode'].astype(str).str.zfill(6)
cnty_gdf = cnty_gdf.to_crs("EPSG:4326")
cnty_gdf
```

```python
sa_pop_log = SATerm(od_data = od_df, dest_data = cnty_gdf, 
                   o_ids = 'origin_geocode', d_ids = 'destination_geocode', 
                   dest_ids = 'geocode', dest_attr = 'pop', 
                   log = True, standard = False)
```

```python
sa_pop_log.cal(-1.0)
cnty_gdf['naive_spop'] = sa_pop_log.dest_df['sat']
```

```python
sa_pop_log.cal(-1.06)
cnty_gdf['106_spop'] = sa_pop_log.dest_df['sat']
```

```python
sa_pop_log.cal(-1.23)
cnty_gdf['123_spop'] = sa_pop_log.dest_df['sat']
```

```python
vmin = min(cnty_gdf['naive_spop'].min(), cnty_gdf['123_spop'].min(), cnty_gdf['106_spop'].min())
vmax = max(cnty_gdf['naive_spop'].max(), cnty_gdf['123_spop'].max(), cnty_gdf['106_spop'].max())
norm = mcolors.Normalize(vmin=vmin, vmax=vmax)
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 5))

# Plot the data without creating individual colorbars
plot1 = cnty_gdf.plot(column='naive_spop', ax=ax1, cmap='Greys', edgecolor='black', lw=0.15, 
                      legend=False, norm=norm)
plot2 = cnty_gdf.plot(column='106_spop', ax=ax2, cmap='Greys', edgecolor='black', lw=0.15, 
                      legend=False, norm=norm)


# Add titles and tick parameters to each subplot
ax1.set_title(r'$\sigma_{naive}$ = -1', fontsize=25)
ax1.tick_params(axis='both', which='major', labelsize=17)

ax2.set_title(r'$\sigma_{io}$ = -1.06', fontsize=25)
ax2.tick_params(axis='both', which='major', labelsize=17)

# ax3.set_title(r'$\sigma_{cd}$ = -1.23', fontsize=25)
# ax3.tick_params(axis='both', which='major', labelsize=17)


# # Create a single colorbar for both subplots
# Save and display the plot
# plt.savefig('pics/sapop_all.png', dpi=300, bbox_inches='tight')
plt.show()
```
