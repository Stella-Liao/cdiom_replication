# Generate points distributed in cluster

```python
# for generating point patterns
from pointpats import PoissonClusterPointProcess, PoissonPointProcess, as_window
from libpysal.cg import Point, Polygon 
```

## One simulation

### Origins & Destinations

```python
# square = Polygon([Point((0, 0)), Point((0, 7000)), Point((7000, 7000)), Point((7000, 0))])
square = Polygon([Point((0, 0)), Point((0, 500)), Point((500, 500)), Point((500, 0))])
squwin = as_window(square)

np.random.seed(1)

#ùëÅ-conditioned CSR:  ùëÅ is fixed
squcsamples = PoissonClusterPointProcess(squwin, 49, 5, 50, 1, asPP=False, conditioning = False)
squpts_cluster = squcsamples.realizations[0]

# Create DataFrame
clu_df = pd.DataFrame(squpts_cluster, columns = ['Lon', 'Lat'])
clu_df = clu_df.assign(Name = np.arange(clu_df.shape[0]))
digits = len(str(49))
clu_df.Name = clu_df.Name.astype(str).str.zfill(digits).apply(lambda x: "{}{}".format('C', x))

# Add attributes, X1
np.random.seed(0)
attr1 = np.random.randint(500, 500001, size=49)
clu_df = clu_df.assign(X1 = attr1)

# Create GeoDataFrame
clu_gdf = gpd.GeoDataFrame(clu_df.copy(), geometry=gpd.points_from_xy(clu_df.Lon, clu_df.Lat))
```

```python
# square = Polygon([Point((0, 0)), Point((0, 7000)), Point((7000, 7000)), Point((7000, 0))])
square = Polygon([Point((0, 0)), Point((0, 500)), Point((500, 500)), Point((500, 0))])

squwin = as_window(square)

np.random.seed(4)

#ùëÅ-conditioned CSR:  ùëÅ is fixed
squsamples = PoissonPointProcess(squwin, 49, 1, conditioning=False, asPP=False)
squpts_random = squsamples.realizations[0]

# Create DataFrame
ran_df = pd.DataFrame(squpts_random, columns = ['Lon', 'Lat'])
ran_df = ran_df.assign(Name = np.arange(ran_df.shape[0]))
digits = len(str(49))
ran_df.Name = ran_df.Name.astype(str).str.zfill(digits).apply(lambda x: "{}{}".format('R', x))

# Add attributes, X1
np.random.seed(4)
attr1 = np.random.randint(500, 500001, size=49)
ran_df = ran_df.assign(X1 = attr1)

# Create GeoDataFrame
ran_gdf = gpd.GeoDataFrame(ran_df.copy(), geometry=gpd.points_from_xy(ran_df.Lon, ran_df.Lat))
```

```python
# Define the square window dimensions
square = Polygon([Point((0, 0)), Point((0, 500)), Point((500, 500)), Point((500, 0))])

# Number of points along one dimension (7x7 grid for 49 points)
n_points = 7

# Generate evenly spaced coordinates
x_coords = np.linspace(0, 500, n_points)
y_coords = np.linspace(0, 500, n_points)

# Create a meshgrid to get all combinations of x and y coordinates
x_grid, y_grid = np.meshgrid(x_coords, y_coords)

# Flatten the grid to get a list of point coordinates
uni_points = np.vstack([x_grid.ravel(), y_grid.ravel()]).T

# Create DataFrame
uni_df = pd.DataFrame(uni_points, columns = ['Lon', 'Lat'])
uni_df = uni_df.assign(Name = np.arange(uni_df.shape[0]))
digits = len(str(49))
uni_df.Name = uni_df.Name.astype(str).str.zfill(digits).apply(lambda x: "{}{}".format('U', x))

# Add attributes, X1
np.random.seed(0)
attr1 = np.random.randint(500, 500001, size=49)
uni_df = uni_df.assign(X1 = attr1)

# Create GeoDataFrame
uni_gdf = gpd.GeoDataFrame(uni_df.copy(), geometry=gpd.points_from_xy(uni_df.Lon, uni_df.Lat))
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols = 3, nrows = 1, figsize = (30,10))

ax1.plot(uni_gdf.Lon.to_numpy(), uni_gdf.Lat.to_numpy(),'o', markeredgecolor='white', color='black', markersize = 14)
ax1.tick_params(axis='both', which='major', labelsize=30)
ax1.set_xlim(-20, 520)
ax1.set_ylim(-20, 520)
ax1.set_title('Uniform', fontsize = 40)

ax2.plot(ran_gdf.Lon.to_numpy(), ran_gdf.Lat.to_numpy(),'o', markeredgecolor='white', color='black', markersize = 14)
ax2.tick_params(axis='both', which='major', labelsize=30)
ax2.set_xlim(-20, 520)
ax2.set_ylim(-20, 520)
ax2.set_title('Random', fontsize = 40)

ax3.plot(clu_gdf.Lon.to_numpy(), clu_gdf.Lat.to_numpy(),'o', markeredgecolor='white', color='black', markersize = 14)
ax3.tick_params(axis='both', which='major', labelsize=30)
ax3.set_xlim(-20, 520)
ax3.set_ylim(-20, 520)
ax3.set_title('Cluster', fontsize = 40)

plt.tight_layout()

plt.savefig("results/simulation_desgin.png", dpi=300)
```
